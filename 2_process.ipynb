```python
import pandas as pd
from tqdm.auto import tqdm
import random
from rdkit import Chem
import numpy as np
import pyarrow.parquet as pq
from collections import defaultdict
```

```python
from pymongo import MongoClient
def get_database():
 
   # Provide the mongodb atlas url to connect python to mongodb using pymongo
   CONNECTION_STRING = "mongodb://anonymous-server.local/belka"
 
   # Create a connection using MongoClient. You can import MongoClient or use pymongo.MongoClient
   client = MongoClient(CONNECTION_STRING)
 
   # Create the database for our example (we will use the same database throughout the tutorial
   return client['belka']
  
   # Get the database
dbname = get_database()
```

```python
dbname.list_collection_names()
collection_name = dbname["bindings_raw_train"]
```

```python
parquet_file = pq.ParquetFile('train.parquet')
for batch in tqdm(parquet_file.iter_batches()):
    batch_df = batch.to_pandas()
    collection_name.insert_many(batch_df.to_dict('records'))
```
