```python
parquet_file = pq.ParquetFile(TEST_MAIN)
results = []
eps=1e-10
from tqdm.auto import tqdm
for batch in tqdm(parquet_file.iter_batches(batch_size=4096)):
    batch = batch.to_pandas()
     # Merge the batch DataFrame with the result DataFrame on 'id_numberic' and 'id'
    merged_df = batch.merge(result_main_island1, left_on='id', right_index=True, how='left')
    merged_df = merged_df.merge(result_main_island1_blend, left_on='id', right_index=True, how='left', suffixes=('_r', '_u'))
    merged_df = merged_df.merge(result_main_island23, left_on='id', right_index=True, how='left')

    def get_score(row):
        # return row[row['protein_name']]
        # return row[row['protein_name']+'_r']
        # # return row[row['protein_name']+'_u']*0.5+row[row['protein_name']+'_r']*0.5
        if row['island']==1: 
            return row['binds_r']
        elif row['island']==0:
            return row['binds_u'] 
        else:
            return row['binds'] 
    merged_df['binds'] = merged_df.apply(get_score, axis=1)
    merged_df = merged_df.drop(['molecule_smiles','protein_name','buildingblock1_smiles','buildingblock2_smiles','buildingblock3_smiles'], axis=1)
    # merged_df = merged_df.drop(['molecule_smiles','protein_name','BRD4','HSA','sEH','buildingblock1_smiles','buildingblock2_smiles','buildingblock3_smiles'], axis=1)
    results.append(merged_df)
```

```python

final_result = pd.concat(results).drop(['island','binds_r','binds_u'],axis=1)
# final_result.to_csv('output_77_s1_mix_random_unique.csv', index=None)
final_result.to_csv('output_final_submission2_fixed_fixed.csv', index=None)
```
